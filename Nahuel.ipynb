```python
import numpy as np
import pandas as pd
```

```python
d2016 = pd.read_csv("insumos/delitos_2016.csv")
d2017 = pd.read_csv("insumos/delitos_2017.csv")
d2018 = pd.read_csv("insumos/delitos_2018.csv")
d2019 = pd.read_csv("insumos/delitos_2019.csv")
```

```python
csvs=[d2016,d2017,d2018,d2019]
```

```python
csvs[1]
```

### Tipos de delitos por año

```python
final=len(csvs)

for e in range(final):
    
    print("Año: ", e+2016)
    base=csvs[e]
    count=base['tipo_delito'].value_counts()
    print(count)
    print("------------------------------------")
```

### Cinco barrios con más delitos por año

```python
for e in range(final):
    
    print("Año: ", e+2016)
    base=csvs[e]
    #print(len(base["barrio"].unique()))
    count=base['barrio'].value_counts()
    print(count.head(5))
    print("------------------------------------")
```

```python
for e in range(final):    
    print("Homicidios por barrio año ", e+2016)
    base=csvs[e][csvs[e].tipo_delito=="Homicidio"]
    #base=base
    count=base['barrio'].value_counts()
    print(count)
    print("Barrios sin homicidios: ", len(csvs[e]["barrio"].unique())-len(base["barrio"].unique()))

    print("------------------------------------")
    print()
```

```python
for e in range(final):    
    print("Robo Con Violencia por barrio Año", e+2016)
    base=csvs[e][csvs[e].tipo_delito=="Robo (con violencia)"]
    #base=base
    count=base['barrio'].value_counts()
    print(count)
    print("Barrios sin robo con violencia: ", len(csvs[e]["barrio"].unique())-len(base["barrio"].unique()))

    print("------------------------------------")
    print()
```

```python
for e in range(final):    
    print("Robo (Con Violencia) por barrio - Año", e+2016)
    base=csvs[e][csvs[e].tipo_delito=="Robo (con violencia)"]
    #base=base
    print("Total de Robos (Con Violencia):", len(base))
    print("Robos con violencia sobre total de delitos:", round((len(base)/len(csvs[e]))*100,2),"%")

    print("------------------------------------")
    print()
```

```python
for e in range(final):    
    print("Año", e+2016)
    base=csvs[e][csvs[e].tipo_delito=="Homicidio"]
    #base=base
    print("Total de homicidios:", len(base))
    print("Homicidios sobre total de delitos:", (len(base)/len(csvs[e]))*100,"%")

    print("------------------------------------")
    print()
```
